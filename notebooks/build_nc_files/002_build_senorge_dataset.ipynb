```python
import xarray as xr
import glob

sd_files = sorted(glob.glob("../data/seNorge/sd_*.nc"))
print(sd_files)

sd_datasets = [xr.open_dataset(f) for f in sd_files]
senorge_sd = xr.concat(sd_datasets, dim="time")

print(senorge_sd)
```

```python
swe_files = sorted(glob.glob("../data/seNorge/swe_*.nc"))
print(swe_files)

swe_datasets = [xr.open_dataset(f) for f in swe_files]
senorge_swe = xr.concat(swe_datasets, dim="time")

print(senorge_swe)
```

```python
senorge_all = xr.merge([senorge_sd, senorge_swe])

print(senorge_all)
```

```python
senorge_all.to_netcdf("../data/processed/senorge_all.nc")
```
