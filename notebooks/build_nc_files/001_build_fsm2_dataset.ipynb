```python
import xarray as xr
import glob

files = sorted(glob.glob("../data/FSM2/fsm2_sd_*.nc"))
print(files)

datasets = [xr.open_dataset(f) for f in files]

fsm2_all = xr.concat(datasets, dim="time")

print(fsm2_all)
```

```python
from pathlib import Path

Path("../data/processed").mkdir(exist_ok=True)

fsm2_all.to_netcdf("../data/processed/fsm2_sd_all.nc")
```
